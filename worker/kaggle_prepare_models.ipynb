This notebook downloads everything **directly from HuggingFace to Kaggle** — nothing goes
through your own computer. After it finishes, save the output as a Kaggle Dataset and
reuse that Dataset in every future worker session (no re-download).

**Settings:** Internet = On. GPU not required for this notebook (it's just downloading).

**After running all cells:** click **Save Version** (top right) → wait for it to finish →
go to the notebook's **Output** tab → **Create Dataset** (or "New Dataset" from the output files).
Name it `openflow-wan22`. That Dataset is what you attach to `kaggle_worker.ipynb`.

```python
!pip install -q huggingface_hub
import os
from huggingface_hub import hf_hub_download

OUT = "/kaggle/working/models"
os.makedirs(OUT, exist_ok=True)

def get(repo, filename, rename=None):
    path = hf_hub_download(repo_id=repo, filename=filename, local_dir=OUT)
    if rename:
        dest = os.path.join(OUT, rename)
        if path != dest:
            os.replace(path, dest)
        print("OK:", rename)
    else:
        print("OK:", os.path.basename(path))
```

```python
# --- T2V UNet GGUF (high + low noise experts) ---
get("QuantStack/Wan2.2-T2V-A14B-GGUF", "HighNoise/Wan2.2-T2V-A14B-HighNoise-Q5_K_M.gguf",
    "wan2.2-t2v-a14b-highnoise-Q5_K_M.gguf")
get("QuantStack/Wan2.2-T2V-A14B-GGUF", "LowNoise/Wan2.2-T2V-A14B-LowNoise-Q5_K_M.gguf",
    "wan2.2-t2v-a14b-lownoise-Q5_K_M.gguf")
```

```python
# --- I2V UNet GGUF (high + low noise experts) ---
get("QuantStack/Wan2.2-I2V-A14B-GGUF", "HighNoise/Wan2.2-I2V-A14B-HighNoise-Q5_K_M.gguf",
    "wan2.2-i2v-a14b-highnoise-Q5_K_M.gguf")
get("QuantStack/Wan2.2-I2V-A14B-GGUF", "LowNoise/Wan2.2-I2V-A14B-LowNoise-Q5_K_M.gguf",
    "wan2.2-i2v-a14b-lownoise-Q5_K_M.gguf")
```
